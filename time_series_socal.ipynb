```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load data
confirmed_raw = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv")

# Clean dataframe and set the index of the current dataframe to the county name
# Before transposing, turn the county name into the column header in the transposed dataset
cases_CA = confirmed_raw[confirmed_raw["Province_State"] == "California"]
cases_CA_T = cases_CA.set_index("Admin2").T.drop(["UID", "iso2", "iso3", "code3", "FIPS", "Province_State", "Country_Region", "Lat", "Long_", "Combined_Key"])
cases_CA_T
```

```python
counties = ["Orange", "Los Angeles", "San Diego"]
plot = cases_CA_T[counties].plot(figsize=(20, 15))
plot.set_title("COVID-19 Cases in Southern California counties")
```
